```python
import pandas as pd
```

```python
death_df = pd.read_csv('20222703 Causes Of Death Clean Output V2.0.csv')
```

```python
death_df
```

```python
countrypop = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_5795797.csv')
```

```python
countrypop
```

```python
# Define a function to retrieve population based on country code and year
def get_population(row):
    country_code = row['Code']
    year = row['Year']
    
    # Check if the country code exists in countrypop DataFrame
    if country_code not in countrypop['Country Code'].values:
        return None  # Return null value if country code doesn't exist
    
    # Check if the year exists in countrypop DataFrame
    if str(year) not in countrypop.columns:
        return None  # Return null value if year doesn't exist
    
    # Use the country code and year to find the population
    population = countrypop.loc[countrypop['Country Code'] == country_code, str(year)].values[0]
    
    return population

# Apply the function to add a new 'Population' column to death_df
death_df['Population'] = death_df.apply(get_population, axis=1)
```

```python
gdp = pd.read_csv('API_NY.GDP.PCAP.CD_DS2_en_csv_v2_5728786.csv')
```

```python
# Define a function to retrieve gdp based on country code and year
def get_gdp(row):
    country_code = row['Code']
    year = row['Year']
    
    # Check if the country code exists in gdp DataFrame
    if country_code not in gdp['Country Code'].values:
        return None  # Return null value if country code doesn't exist
    
    # Check if the year exists in gdp DataFrame
    if str(year) not in gdp.columns:
        return None  # Return null value if year doesn't exist
    
    # Use the country code and year to find the gdp
    gdpvalue = gdp.loc[gdp['Country Code'] == country_code, str(year)].values[0]
    
    return gdpvalue

# Apply the function to add a new 'Gdp' column to death_df
death_df['Gdp'] = death_df.apply(get_gdp, axis=1)
```
